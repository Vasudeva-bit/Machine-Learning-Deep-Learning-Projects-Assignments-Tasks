### DeepAREstimator (using gluonts library-MXNet)

* Problem identified and Solved by DeepAR paper:
   * The very important usecase of DeepAR model is to forecast multiple time series. The major problem identified is, traditional estimators for multiple series forecasting couldn't accurately forecast multiple time series whose magnitudes differ widely (multiple time series of different scales or power law distribution). The real example of such, is Amazon sell different items among millions them and this produce a multiple time series with differnt scales. (Direct example from DeepAR)
   * The DeepAR though jointly learns from multiple time series could effectively handle hetrogeneous nature of time series for following reasons. 

* DeepAR outperforms excisting models:
   * The DeepAR model could perform well despite having magnitudes of multiple time series that differ widely, is due to its network.
   * The model distribution consists of product of two likelihood factors. The likelihood functions are Negative Binomial and Guassian distribution.
   * The DeepAR is a sequence to sequence model with encoder and decoder. This architecture allows estimator to forecast multiple time series data with improved accuracy.
   * As the model learns seasonal behavior and dependencies on given covariates across time series, the manual feature engineering is minimized though for complex group-dependent behavior. 
   * By learning jointly from similar items or time series, DeepAR is able to provide forecasts for items with little or no history at all, where traditional single time series forecasting methods fail.
   * DeepAR address this issue by dividing the autoregressive inputs
by an item-dependent scale factor, and conversely multiplying the scale-dependent likelihood parameters by the same factor

* The Advantage of RNN units in the DeepAR Estimator:
   * Unlike statistical models, RNN units such as LSTMs or GRUs use the prediction of current time step and feed it for the predict value at next time step. This allows RNN units of DeepAR model to accurately forecast multiple time series with longer forecast horizons.

```python
freq = '15min'
start_train = pd.Timestamp('2011-01-01 00:15:00', freq = freq)
start_test = pd.Timestamp('2014-11-07 05:30:00', freq=freq)
prediction_length = 672
```

```python
estimator = DeepAREstimator(freq=freq,
                            context_length=672,
                            prediction_length=prediction_length,
                            use_feat_static_cat=True,
                            cardinality=[i],
                            num_layers=2,
                            num_cells=32,
                            cell_type='lstm',
                            trainer=Trainer(epochs=4)) #instantiate the model
```

```python
from gluonts.dataset.common import ListDataset
from gluonts.dataset.field_names import FieldName

train_ds = ListDataset([
    {
        FieldName.TARGET: target,
        FieldName.START: start_train,
        FieldName.FEAT_STATIC_CAT: fsc
    }
    for (target, fsc) in zip(df_test[0:7],
                             ts_code[0:7].reshape(-1,1))
    
], freq=freq)

test_ds = ListDataset([
    {
        FieldName.TARGET: target,
     FieldName.START: start_test,
     FieldName.FEAT_STATIC_CAT: fsc
    }
    for(target, fsc) in zip(df_test[0:7],
                            ts_code[0:7].reshape(-1,1))
], freq=freq)
```

```python
next(iter(train_ds))
```

```python
predictor = estimator.train(training_data=train_ds)
```

```python
from gluonts.evaluation.backtest import make_evaluation_predictions
```

```python
forecast_it, ts_it = make_evaluation_predictions(
    dataset=test_ds,
    predictor=predictor,
    num_samples=100,
)
```

```python
from tqdm.autonotebook import tqdm
```

```python
print("Obtaining time series conditioning values...")
tss = list(tqdm(ts_it, total=len(df_test)))
print("Obtaining time series predictions...")
forecasts = list(tqdm(forecast_it, total=len(df_test)))
```

```python
def plot_prob_forecasts(ts_entry, forecast_entry):
  plot_length = prediction_length
  prediction_intervals = (80.0, 95.0)
  legend = ["observations", "median prediction"] + [f"{k}% prediction interval" for k in prediction_intervals[::-1]]

  fig, ax = plt.subplots(1, 1, figsize=(10, 7))
  ts_entry[-plot_length:].plot(ax=ax)
  forecast_entry.plot(prediction_intervals=prediction_intervals, color='g')
  plt.grid(which="both")
  plt.legend(legend, loc="upper left")
  plt.show()
```

```python
for i in tqdm(range(7)):
  ts_entry = tss[i]
  forecast_entry = forecasts[i]
  plot_prob_forecasts(ts_entry, forecast_entry)
```

```python
from gluonts.evaluation import Evaluator
evaluator = Evaluator(quantiles=[0.1, 0.5, 0.9])
agg_metrics, item_metrics = evaluator(iter(tss), iter(forecasts), num_series=len(df_test[0:7]))
```
